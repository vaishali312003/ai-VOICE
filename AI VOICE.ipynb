#### JUST TO MAKE OUR BOT TO CONVERT FROM TEXT TO AUDIO 

```python
import gradio as gr
from gtts import gTTS

import os


def generate_voice(text):
    
    tts = gTTS(text=text, lang='en', slow=False)

    tts.save("output.mp3")

   
    os.system("start output.mp3")

   
    return "Voice generated and played!"


iface = gr.Interface(
    fn=generate_voice,
    inputs="text",
    outputs="text"
)

iface.launch()
```
